```python
import pandas as pd
import os
```

```python
list_png = []
path = '/home/dell/Documentos/OCT/chest_xray'

for root, dirs, files in os.walk(path):
    for file in files:
        if(file.endswith(".jpeg")):
            png_file = os.path.join(root,file)
            list_png.append(png_file)
            
```

```python

dir_normal = "/home/dell/Documentos/OCT/normal_images"
dir_virus = "/home/dell/Documentos/OCT/p_virus_images"
dir_bac = "/home/dell/Documentos/OCT/p_bac_images"

# Criando diret√≥rios novos

for path_lb in [dir_normal,dir_virus,dir_bac]:
    try:
        os.makedirs(path_lb)
    except OSError:
        print ("Creation of the directory %s failed" % path_lb)
    else:
        print ("Successfully created the directory %s" % path_lb)
```

```python
virus = []
bac = []
normal = []

for img in list_png:
    if "virus" in img:
        virus.append(img)
    elif "bac" in img:
        bac.append(img)
    else:
        normal.append(img)
```
