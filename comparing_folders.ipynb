```python
import os
import cv2

root = "2_Raw_Seg/"
count_files = 0
bad_size = 0

for path, subdirs, files in os.walk(root):
    for name in files:
        file_path = os.path.join(path, name)

        if file_path.find("BIMCV") > 0 :
            print("not valid")
        
        elif file_path.find("Not normal") > 0 :
            print("not valid")
        
        else:
            img = cv2.imread(file_path, 0)
            w, h = img.shape
            if w < 250:
                bad_size += 1
            else:
                count_files += 1


count_files
```
